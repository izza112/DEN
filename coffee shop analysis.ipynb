```python
data['total_sales'] = data['transaction_qty'] * data['unit_price']

product_sales = data.groupby('product_detail')['total_sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 8))
product_sales.head(10).plot(kind='bar', color='violet')
plt.xlabel('Product Detail')
plt.ylabel('Total Sales')
plt.title('10 most profitable Product')
plt.xticks(rotation=90)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
product_sales.tail(10).plot(kind='bar', color='violet')
plt.xlabel('Product Detail')
plt.ylabel('Total Sales')
plt.title('10 least profitable Product')
plt.xticks(rotation=90)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
x=data[['transaction_qty','unit_price']]
y=data['total_sales']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
print("Mean Squared Error:", mean_squared_error(y_test, prediction))
print("R-squared:", r2_score(y_test, prediction))
```

```python
low_performance_threshold=data['total_sales'].quantile(0.05)
low_performance_product=data[data['total_sales']<low_performance_threshold]

print(low_performance_product['product_detail'].value_counts())
```

```python
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd


data['transaction_date'] = pd.to_datetime(data['transaction_date'], format='%d-%m-%Y')
daily_sales = data.groupby('transaction_date')['total_sales'].sum()

model = ARIMA(daily_sales, order=(1, 1, 1))
model_fit = model.fit()
```

```python
forecast=model_fit.forecast(steps=30)
plt.figure(figsize=(12, 8))
plt.plot(daily_sales.index, daily_sales.values, label='Actual Sales')
plt.plot(forecast,label='Forecasted sales',color='red')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Daily Sales Forecast')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
import seaborn as sns

features=data[['transaction_qty','unit_price']]
kmeans=KMeans(n_clusters=3,random_state=42)
data['cluster']=kmeans.fit_predict(features)
```

```python
sns.scatterplot(x='transaction_qty',y='unit_price',hue='cluster',data=data)
plt.title('Customer Segmentation')
plt.show()
```
